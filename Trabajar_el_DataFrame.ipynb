```
casosDiaPorFecha = covidColombia2020_06_19PorFecha.Caso
recuperados = covidColombia2020_06_19[covidColombia2020_06_19['Tipo_recuperacion']!='No recuperado']
recuperadosDiaPorFecha = recuperados.groupby('Fecha Not').count()
recuperados = recuperados['Fecha Not'].drop_duplicates()
recuperadosDiaPorFecha = recuperadosDiaPorFecha['Caso']
plt.figure(figsize=(23,11))
ax = plt.axes()
ax.set_xlabel('\nTiempo',size=25,color='#4bb4f2')
ax.set_ylabel('Contagiados\n', size=25,color='#4bb4f2') 
plt.plot_date(fechas,casosDiaPorFecha, linestyle='solid')
plt.plot_date(recuperados,recuperadosDiaPorFecha, linestyle='solid',color="green")
plt.legend(['Contagios','Recuperados'],fontsize=20)
plt.show()
```
